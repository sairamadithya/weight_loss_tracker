```python
def exercise_calculator(walk,run,push,plank,squat,pull,dip,cycle,bench,leg,cross):
    c1= (5*walk)+(9*run)+(7*push)+(4*plank)+(8*squat)+(1*pull)+(10*dip)+(7*cycle)+(5*curl)+(40*bench)+(4*leg)+(16*cross)
    return c1   
print("walking, running, pushups, planks, squats, pull-ups, dips, cycling, bicep curl, bench press, leg press, cross-fit")
print("enter either the number of minutes or reps of the corresponding activity or enter 0 if not done")
print("enter the minutes of walking")
walk= int(input())
print(" enter the minutes of running")
run=int(input())
print("enter the minutes of pushup")
push= int(input())
print("enter the minutes of planks")
plank= int(input())
print("enter the minutes of squats")
squat= int(input())
print("enter the reps of pullups")
pull= int(input())
print("enter the sets of dips")
dip= int(input())
print("enter the minutes of cycling")
cycle= int(input())
print("enter the sets of bicep curl")
curl= int(input())
print("enter the sets of bench press")
bench= int(input())
print("enter the minutes of leg press")
leg= int(input())
print("enter the minutes of crossfit")
cross= int(input())
calories= exercise_calculator(walk,run,push,plank,squat,pull,dip,cycle,bench,leg,cross)
print("the total number of calories burnt is",calories)



    
```
